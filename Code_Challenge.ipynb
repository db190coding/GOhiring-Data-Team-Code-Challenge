```python
# remove duplicates from df_jobs
df_jobs = df_jobs.drop_duplicates()
```

```python
# join df_jobs and df locations
df_jobs_locations = pd.merge(df_jobs, df_locations, left_on="zip", right_on="zip_code", how="inner")
```

```python
# join all df to df_all
df_all = pd.merge(df_jobs_locations, df_companies, left_on="company_id", right_on="Company ID", how="inner")
```

```python
# check unique values
for column in df_all.columns:
  print(df_all[column].unique())
```

```python
# correct spelling mistakes
df_all["state"].replace("canceled", "cancelled", inplace=True)
df_all["state"].replace("osted", "posted", inplace=True)
```

```python
# check unique values to check if spelling mistakes are corrected
for column in df_all.columns:
  print(df_all[column].unique())
```

```python
# drop duplicate columns due to join
df_all.drop("zip_code", axis=1, inplace=True)
df_all.drop("Company ID", axis=1, inplace=True)
# replace spaces with underscores
df_all.columns = df_all.columns.str.replace(" ", "_")
df_all.columns = df_all.columns.str.lower()
```

```python
# show df_all
df_all
```

```python
# Question1: What location has the most jobs that are either posted or expired?
df_only_posted_expired = df_all[df_all["state"].isin(["posted", "expired"])]
location_posted_expired = df_only_posted_expired["location"].value_counts()
location_posted_expired
```

```python
# Add month column (preprocessing for Question 2)
df_all["posted_at_month"] = df_all["posted_at"].dt.month
```

```python
# Show modified df_all
df_all
```

```python
# Question 2: What month had the most cancelled jobs?
df_only_cancelled = df_all[df_all["state"] == "cancelled"]
month_cancelled = df_only_cancelled["posted_at_month"].value_counts()
month_cancelled
```

```python
# Filter to df_all to state == posted:
df_only_posted = df_all[df_all["state"] == "posted"]
# count values by comany
company_posted = df_only_posted["company_name"].value_counts()
```

```python
# show result
company_posted
```

```python
# convert to dataframe
df_company_posted = pd.DataFrame(company_posted)
```

```python
# reset index
df_company_posted.reset_index(inplace = True)
```

```python
# show dataframe
df_company_posted
```

```python
# join with df companies to include the Number of employees
df_posted_empoloyees_ratio = pd.merge(df_company_posted, df_companies, left_on="company_name", right_on="Company Name", how="inner")
```

```python
# show result dataframe
df_posted_empoloyees_ratio
```

```python
# calculate the ratio of posted jobs to employee count
df_posted_empoloyees_ratio["ratio"] = df_posted_empoloyees_ratio["count"] / df_posted_empoloyees_ratio["Number of Employees"]
```

```python
# show the modified datframe
df_posted_empoloyees_ratio
```

```python
# Question 3: Which company has the highest ratio of posted jobs to employee count?
# retrieve the column with the max ratio
max_index = df_posted_empoloyees_ratio["ratio"].idxmax()
max_ratio = df_posted_empoloyees_ratio.loc[max_index]
max_ratio
```

```python
# Create df with necessary columns for the chart
df_chart =  df_only_posted[["posted_at","company_name"]]
# convert data to period
df_chart["posted_at"] = df_chart["posted_at"].dt.to_period("M")
```

```python
# show df_chart
df_chart
```

```python
# Question 7: (For Data Analyst Role) Provide a visualization showing the trend of the number of active jobs over time. Also include the company name data in this visualization.
import matplotlib.pyplot as plt

grouped = df_chart.groupby(["posted_at", "company_name"]).size().unstack(fill_value=0)

# Plot the stacked column chart
grouped.plot(kind="bar", stacked=True, figsize=(12, 6))
plt.xlabel("Year-Month")
plt.ylabel("Count")
plt.title("Count of Posted Dates by Company (Monthly)")
plt.xticks(rotation=45)
plt.legend(title="Company")
plt.show()
```

```python
'''SQL Questions:

Question 4:

-- Locations Dimension Table
CREATE TABLE locations (
    zip_code VARCHAR2(10) PRIMARY KEY,
    location VARCHAR2(100) NOT NULL
);

-- Companies Dimension Table
CREATE TABLE companies (
    company_id NUMBER PRIMARY KEY,
    company_name VARCHAR2(100) NOT NULL,
    establishment_date DATE,
    number_of_employees NUMBER
);

-- Jobs Fact Table
CREATE TABLE jobs (
    id NUMBER PRIMARY KEY,
    posted_at TIMESTAMP,
    state VARCHAR2(50),
    zip_code VARCHAR2(10),
    price NUMBER(10, 2), -- or use DECIMAL if preferred
    company_id NUMBER,
    FOREIGN KEY (zip_code) REFERENCES locations(zip_code),
    FOREIGN KEY (company_id) REFERENCES companies(company_id)
);

Question 5:

SELECT
    company_name,
    RANK() OVER (PARTITION BY c.company_id ORDER BY j.posted_at) AS job_rank,
    j.id AS job_id,
    j.posted_at,
    j.state,
    j.zip_code,
    j.price
FROM
    companies c
INNER JOIN
    jobs j ON c.company_id = j.company_id
ORDER BY
    c.company_name, j.posted_at;

Question 6: We would need to add ann additonal column "expired_at" to the fact table jobs. Then we would be able to calculate the duration from posted to expired:

ALTER TABLE jobs ADD expired_at TIMESTAMP;
```
